```python
#ell_min = 200
def Filter_Map3(Map, N, N_mask, ell_min):
    ones = np.ones(N) # N is the size of the map
    inds  = (np.arange(N)+.5 - N/2.) 

    X = np.outer(ones,inds)
    Y = np.transpose(X)
    R = np.sqrt(X**2. + Y**2.)  ## angles realative to 1 degrees  
    
    ## make a mask
    mask  = np.ones([N,N])
    mask[np.where(np.abs(X) < N_mask)]  = 0
    
    # apply the filter in Fourier space
    FMap = np.fft.fftshift(np.fft.ifft2(np.fft.fftshift(Map)))
    ft_map = np.fft.fft2(Map)
    ft_map_shifted = np.fft.fftshift(ft_map)
    #freqs = np.fft.fftfreq(Map.shape[0]) * Map.shape[0]
    freqs = np.fft.fftfreq(Map.shape[0])[:, np.newaxis]
    k = 2 * np.pi * freqs
    ell = k * Map.shape[0] / (2 * np.pi)
    #ell = np.abs(freqs)
    filters = np.where(ell > ell_min, 1, 0)
    FMap_filtered = FMap * mask * filters
    Map_filtered = np.real(np.fft.fftshift(np.fft.fft2(np.fft.fftshift(FMap_filtered))))

    
    # return the output
    return Map_filtered


filtered_map = Filter_Map3(appodized_map, N, 10, ell_min=200)
p = Plot_CMB_Map(filtered_map, c_min, c_max, N, N)
binned_ell_dat_cut, binned_spectrum_dat_cut = \
    calculate_2d_spectrum(filtered_map, delta_ell, ell_max, pix_size, N)
plt.semilogy(binned_ell_theory,binned_spectrum_theory*\
                 binned_ell_theory * (binned_ell_theory+1.)/2. / np.pi, label='theory')
plt.semilogy(binned_ell_dat_cut,binned_spectrum_dat_cut*\
                binned_ell_dat_cut * (binned_ell_dat_cut+1.)/2. / np.pi, label='data')

leg = plt.legend()
leg.draw_frame(False)
plt.semilogy(ell,DlTT)
plt.ylabel('$D_{\ell}$ [$\mu$K$^2$]')
plt.xlabel('$\ell$')
plt.show()
#Go through step by step to debug
```

```python
#ell_min = 200
def Filter_Map2(Map, N, N_mask, ell_min):
    N = int(N)
    
    # set up a 2D ell coordinate system
    ones = np.ones(N)
    inds = (np.arange(N) + 0.5 - N/2.) / (N - 1.)
    kX = np.outer(ones, inds) / (pix_size/60. * np.pi/180.)
    kY = np.transpose(kX)
    K = np.sqrt(kX**2. + kY**2.)
    ell_scale_factor = 2. * np.pi 
    ell2d = K * ell_scale_factor
    
    # make a mask
    mask = np.ones([N, N])
    mask[np.where(ell2d < ell_min)] = 0
    
    # apply the filter in Fourier space
    FMap = np.fft.fftshift(np.fft.ifft2(np.fft.fftshift(Map)))
    FMap_filtered = FMap * mask
    Map_filtered = np.real(np.fft.fftshift(np.fft.fft2(np.fft.fftshift(FMap_filtered))))

    
    # return the output
    return Map_filtered


filtered_map = Filter_Map2(appodized_map, N, 10, ell_min=200)
p = Plot_CMB_Map(filtered_map, c_min, c_max, N, N)
binned_ell_dat_cut, binned_spectrum_dat_cut = \
    calculate_2d_spectrum(filtered_map, delta_ell, ell_max, pix_size, N)
plt.semilogy(binned_ell_theory,binned_spectrum_theory*\
                 binned_ell_theory * (binned_ell_theory+1.)/2. / np.pi, label='theory')
plt.semilogy(binned_ell_dat_cut,binned_spectrum_dat_cut*\
                binned_ell_dat_cut * (binned_ell_dat_cut+1.)/2. / np.pi, label='data')

leg = plt.legend()
leg.draw_frame(False)
plt.semilogy(ell,DlTT)
plt.ylabel('$D_{\ell}$ [$\mu$K$^2$]')
plt.xlabel('$\ell$')
plt.show()
#Go through step by step to debug
```
